```
from simpletransformers.classification import ClassificationModel
albert_model = ClassificationModel('albert', 'albert-base-v1', num_labels=3, args={'learning_rate':1.5e-5, 'num_train_epochs': 25, 'reprocess_input_data': True, 'overwrite_output_dir': True, 'save_model_every_epoch' : False})
roberta_model = ClassificationModel('roberta', 'roberta-base', num_labels=3, args={'learning_rate':1.5e-5, 'num_train_epochs': 25, 'reprocess_input_data': True, 'overwrite_output_dir': True, 'save_model_every_epoch' : False})
bert_model =  ClassificationModel('bert', 'bert-base-uncased', num_labels=3, args={'learning_rate':1.5e-5, 'num_train_epochs': 25, 'reprocess_input_data': True, 'overwrite_output_dir': True, 'save_model_every_epoch' : False})
xlnet_model = ClassificationModel('xlnet', 'xlnet-base-cased', num_labels=3, args={'learning_rate':1.5e-5, 'num_train_epochs': 25, 'reprocess_input_data': True, 'output_dir' : '/content/drive/My Drive/ExperimentsBERT/model/xlnet_correct_data/','overwrite_output_dir': True, 'save_model_every_epoch' : False})
```

```
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix, classification_report
albert_model.train_model(train_df, eval_df=eval_df)
roberta_model.train_model(train_df, eval_df=eval_df)
bert_model.train_model(train_df, eval_df=eval_df)
xlnet_model.train_model(train_df, eval_df=df3)



def f1_multiclass(labels, preds):
    return f1_score(labels, preds, average='micro')

result, model_outputs, wrong_predictions = xlnet_model.eval_model(eval_df=df3, f1=f1_multiclass, acc=accuracy_score, confusion_matrix = confusion_matrix, cr = classification_report)
print(result)
print(result['cr'])

  
result, model_outputs, wrong_predictions = albert_model.eval_model(eval_df, f1=f1_multiclass, acc=accuracy_score, confusion_matrix = confusion_matrix, cr = classification_report)
print(result)
print(result['cr'])


result, model_outputs, wrong_predictions = roberta_model.eval_model(eval_df, f1=f1_multiclass, acc=accuracy_score, confusion_matrix = confusion_matrix, cr = classification_report)
print(result)
print(result['cr'])

result, model_outputs, wrong_predictions = bert_model.eval_model(eval_df, f1=f1_multiclass, acc=accuracy_score, confusion_matrix = confusion_matrix, cr = classification_report)
print(result)
print(result['cr'])


```
