```python
from simpletransformers.classification import ClassificationModel

xlnet_model = ClassificationModel('xlnet', 'xlnet-base-cased', num_labels=4, args={'learning_rate':1.5e-5, 'num_train_epochs': 25, 'reprocess_input_data': True, 'output_dir': '/content/drive/My Drive/ExperimentsBERT/component_corpus/component_models/xlnet/', 'overwrite_output_dir': True, 'save_model_every_epoch' : False})
```

```python
from simpletransformers.classification import ClassificationModel
model_path = '/content/drive/My Drive/ExperimentsBERT/component_corpus/component_models/roberta/'
roberta_model = ClassificationModel('roberta','roberta-base', num_labels=4, args={'learning_rate':1.5e-5, 'num_train_epochs': 25, 'reprocess_input_data': True, 'output_dir': '/content/drive/My Drive/ExperimentsBERT/component_corpus/component_models/roberta/modified/', 'overwrite_output_dir': True, 'save_model_every_epoch' : False})
```

```python
#Train and test using Finetuned XLNet
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix, classification_report

xlnet_model.train_model(train_df, eval_df=test_df)

def f1_multiclass(labels, preds):
    return f1_score(labels, preds, average='micro')

result, model_outputs, wrong_predictions = xlnet_model.eval_model(eval_df=test_df, f1=f1_multiclass, acc=accuracy_score, confusion_matrix = confusion_matrix, cr = classification_report)
print(result)
print(result['cr'])
```
